## Reshaping, shaping, sqeezing and unsqueezing

```python
y = torch.arange(1.,10.)
y,y.shape
```

```python
y_reshaped = y.reshape(9,1)
y_reshaped
```

```python
z = y.view(3,3)
z
```

```python
x_stacked = torch.stack([y,y,y,y], dim =1)
x_stacked
```

```python
x_squeezed = y.squeeze()
x_squeezed,x_squeezed.shape
```

```python
x_unsqeezed = x_squeezed.unsqueeze(dim=0)
x_unsqeezed,x_unsqeezed.shape
#
```

```python
x_original = torch.rand(size=(224,224,3))

x_permuted = x_original.permute(2,0,1)
print(f"previous shape = {x_original.shape}")
print(f"permuted shape = {x_permuted.shape}")
```
